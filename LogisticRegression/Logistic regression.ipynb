```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
%matplotlib inline
digits=load_digits()
```

```python
print("Image data shape: ",digits.data.shape)
print("Label data shape: ",digits.target.shape)
```

```python
import numpy as np
```

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:9],digits.target.data[0:9])):
    plt.subplot(1,9,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title("Training%i\n"%label,fontsize=20)
```

```python
np.info(digits.target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.34, random_state=2)
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
print(model.predict(X_test[0].reshape(1,-1)))
```

```python
predictions=model.predict(X_test)
predictions
```

```python
score=model.score(X_test,y_test)
```

```python
score
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import metrics
```

```python
cm=metrics.confusion_matrix(y_test,predictions)
cm
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt=".3f",linewidths=.5,square=True,cmap='Blues_r');
plt.ylabel('Actual Value')
plt.xlabel("Predicted Values")
all_sample_title='Accuracy Score : {0}'.format(score)
plt.title(all_sample_title, size=15)
```

```python
index=0
CI=[]
for predict, actual in zip(predictions,y_test):
    if(predict!=actual):
        CI.append(index)
    index+=1
plt.figure(figsize=(20,3))
for plotIndex,wrong in enumerate(CI[0:4]):
    plt.subplot(1,4,plotIndex+1)
    plt.imshow(np.reshape(X_test[wrong], (8,8)),cmap=plt.cm.gray)
    plt.title("Predicted: {}, Actual{}".format(predictions[wrong],y_test[wrong]), fontsize=20)
```
